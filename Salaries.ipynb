Split Column: The str.split() method is used to split the EmployeeName column into two parts. The first argument to split() is the delimiter (a space in this case). The second argument (1) specifies that the split should happen only at the first occurrence of the delimiter. expand=True ensures that the split parts are expanded into separate columns.

```python
# Step 3: Split the 'EmployeeName' Column
salaries_df[['FirstName', 'LastName']] = salaries_df['EmployeeName'].str.split(n=1, expand=True)

# Step 4: Create a Boolean Mask for First Name 'John'
john_first_name_mask = salaries_df['FirstName'] == 'JOHN'

# Step 5: Count the Rows
john_first_name_count = john_first_name_mask.sum()

# Step 6: Display the Result
print(f"Number of employees whose first name is 'John': {john_first_name_count}")
```

### 8. Append a new row to the DataFrame to represent a made up employee (you can use the pd.Concat() method for this)
- He/she has no benefits, so set this attribute to not a number using np.nan
- Make up the other column values, but make sure they are standardised to the same convention as the other values in that column

#### step:1 Creating a new row  as a DataFrame including a column for benefits set to np.nan 

Why Do I need to create a new row as a Dataframe first before adding it to my existing Dataframe?

1. Consistency with existing DataFrame structure
- The existing Dataframe salaries_df has a specific structure with named columns.
- Creating a new row as a Dataframe, ensures that the new row has the same column structure as the Dataframe 

2. Handling Data Types:

- DataFrames in Pandas handle various data types (strings, integers, floats, etc.). By creating the new row as a DataFrame, you ensure that each value is appropriately typed according to the corresponding column in the existing DataFrame.
- For example, setting Benefits to np.nan properly handles the missing value as a float, which is a common representation for missing data in Pandas.

3. Scalability and Flexibility

- If you need to add multiple new rows, it is easier to create a DataFrame with those rows and concatenate it with the existing DataFrame.

4. Ease of Concatenation

- <b>pd.concat()</b> is designed to work with DataFrames. Concatenating two DataFrames along the row axis (axis 0) is straightforward and efficient.
- Using a DataFrame for the new row ensures compatibility with pd.concat(), avoiding potential issues that might arise from concatenating other data structures.

Based on the provided column names and data types, we'll create a new row for a made-up employee. The new employee will have the necessary attributes filled in while setting Benefits, Notes, and Status to np.nan since they have no values.

```python
import numpy as np

new_employee = pd.DataFrame({
    'Id': ['12345'],                    # Example ID
    'EmployeeName': ['Jane Smith'],     # Example Employee Name
    'JobTitle': ['Software Engineer'],  # Example Job Title
    'BasePay': [90000.0],               # Example Base Pay
    'OvertimePay': [5000.0],            # Example Overtime Pay
    'OtherPay': [3000.0],               # Example Other Pay
    'Benefits': [np.nan],               # No benefits
    'TotalPay': [98000.0],              # BasePay + OvertimePay + OtherPay
    'TotalPayBenefits': [98000.0],      # TotalPay + Benefits (Benefits are nan, hence ignored)
    'Year': [2024],                     # Example Year
    'Notes': [np.nan],                  # No notes
    'Agency': ['Tech Department'],      # Example Agency
    'Status': [np.nan],                 # No status
    'FirstName': ['Jane'],              # Example First Name
    'LastName': ['Smith'],              # Example Last Name
    'last_updated': ['2024-07-03'],     # Example last updated date
    'Overtime_Ratio': [0.055],          # Example Overtime Ratio (OvertimePay/BasePay)
    'SumPay': [98000.0]                 # Example SumPay (TotalPay)
})
```

### Concatenate the new row to the DataFrame:

```python
df = pd.concat([salaries_df, new_employee], ignore_index=True)
```

### Write the updated DataFrame back to the CSV file:

Writing the updated DataFrame back to the CSV file is necessary to persist the changes made to the DataFrame. Imagine you have an employee database stored in a CSV file. If you don't write the updated DataFrame back to the CSV file, the next time you read the CSV file, the new employee's information won't be there. Writing the updated DataFrame back to the CSV file ensures that the new employee's information is permanently saved.
