### Step 8. Sort the teams by Red Cards, then to Yellow Cards
We want to highlight the teams that are been involved in more severe incidents during matches 

```python
# discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=[False, False], inplace=True)
```

(SettingWithCopyWarning) indicates that a slice of DataFrame Euro12 is modified directly, which can lead to unexpected behavior. To avoid this warning, we can modify the code to create a copy of the relevant DataFrame slice before sorting.

```python
 #Create a new DataFrame with relevant columns and make a copy to avoid SettingWithCopyWarning
discipline = euro12[['Team', 'Red Cards', 'Yellow Cards']].copy()
```

```python
discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=[False, False], inplace=True)
```

```python
# Display the sorted DataFrame
print(discipline)
```

```python
discipline.head(10)
```

#### Notes :
Explaining Sorting Order: ascending=[False, False]:

When we sort a DataFrame, we can specify whether we want the values to be arranged in ascending or descending order.By default, pandas sorts in ascending order (from smallest to largest).

Descending Order:
To sort in descending order (from largest to smallest), we set ascending=False.This means that higher values will appear first in the sorted result.
