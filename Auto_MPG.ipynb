### Step 7. Oops, there is a column missing, called owners. Create a random number Series from 15,000 to 73,000.

```python
import numpy as np
```

To generate a random number for each row, we first need to know how many rows are in the DataFrame. This is done using result.shape[0], where shape[0] returns the number of rows.We know how many rows are in the datashape(look above) but we could also use shape[] to extract this information and use it whilst adding a new column with new random numbers as values :

```python
num_rows = result.shape[0]
```

We want to create random integers between 15,000 and 73,000 for each row. We use numpy's randint function to generate these numbers

```python
random_numbers = np.random.randint(15000, 73001, size= num_rows)
```

- 15000: The lower bound (inclusive).
- 73001: The upper bound (exclusive). We use 73001 because randint does not include the upper bound, so this ensures we get numbers up to 73,000.
- size=num_rows: The number of random integers to generate, equal to the number of rows in the DataFrame.

We add these random numbers as a new column named owners to the DataFrame. This is done by assigning the random_numbers array to result['owners']:

```python
result['owners'] = random_numbers
```
